```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('seaborn-v0_8-notebook')
```

```python
eps = 1.0
eps_decs = [1., 0.99999, 0.99998, 0.999975, 0.99997, 0.999965]
labels = ['none', 'very slow', 'slow', 'medium', 'fast', 'faster']
# alpha = 0.1
# alpha_dec = 0.99999
```

```python
plt.figure(figsize=(20, 8))
N = int(4e5)
x = np.arange(N)
plot = plt.semilogy
for eps_dec, label in zip(eps_decs, labels):
    plot(x, eps * eps_dec ** x, label=label + ' ={}'.format(eps_dec))
# plot(x, alpha * alpha_dec ** x, label='alpha')
plt.xlabel('Episodes')
plt.ylabel('Value')
plt.legend(fontsize=22)
plt.grid()
plt.show()
```

```python
eps0 = 1.0
eps_d = 0.99999
eps_f = 5e-5
n_epochs = lambda eps_d: np.ceil((np.log(eps_f) - np.log(eps0)) / np.log(eps_d)).astype(int)
```

```python
plt.figure(figsize=(20, 8))
x = np.array([0.999965, 0.99997, 0.999975, 0.99998, 0.999985, 0.99999, 0.999995])
y = n_epochs(x)
plt.semilogy(x, y)
plt.xlabel('Decay')
plt.ylabel(f'Number of epochs to get to {eps_f:.2f}')
plt.grid()
plt.show()
```
